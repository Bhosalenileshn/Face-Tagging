```python
import os
import matplotlib.pyplot as plt
import matplotlib as m
```

```python
matched_pairs = {
    "group_1": ["image1.jpg", "image4.jpg"],
    "group_2": ["image2.jpg"],
    "group_3": ["image6.jpg"]
}
```

```python
folder_path = "images"

fig, axes = plt.subplots(nrows=len(matched_pairs), ncols=2)

for ax, (group, images) in zip(axes, matched_pairs.items()):
    for a, image in zip(ax, images):
        img = plt.imread(os.path.join(folder_path, image))
        a.imshow(img)
fig.delaxes(axes[1][1])
fig.delaxes(axes[2][1])
```
